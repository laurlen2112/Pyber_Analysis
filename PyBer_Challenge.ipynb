### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os

# File to Load (updated)
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type 
#use the groupby()to create a Series with type of city as the index, then count()"ride_id" column
rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_by_type
```

```python
# 2. Get the total drivers for each city type
#use the groupby()to create a Series with type of city as the index, then sum()"driver_count" column
drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
# use the groupby() create a Series with type of city as the index,then sum() the "fare" column.
fares_by_type = pyber_data_df.groupby("type").sum()["fare"]
fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fares_by_type = pyber_data_df.groupby("type").mean()["fare"]
avg_fares_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = fares_by_type/ drivers_by_type 
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides": rides_by_type, "Total Drivers" : drivers_by_type,
                               "Total Fares": fares_by_type, "Average Fare per Ride": avg_fares_by_type, 
                                "Average Fare per Driver": avg_fare_per_driver})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)

pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format) 

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
                                                  
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)                               

pyber_summary_df                                      
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

In Step 1, create a new DataFrame with multiple indices using the 
groupby() function on the "type" and "date" columns of the pyber_data_df DataFrame, 
then apply the sum() method on the "fare" column to show 
the total fare amount for each date.

```python
# 1. Read the merged DataFrame

pyber_data_df.describe()
pyber_data_df.shape
pyber_data_df.head()
pyber_data_df.info()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

grouped_df = pyber_data_df
grouped_df = grouped_df.groupby(["type", "date"]).sum()[["fare"]]
grouped_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

grouped_df = grouped_df.reset_index()
grouped_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
grouped_df = grouped_df.pivot(index = "date", columns = "type", values= "fare")
grouped_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_loc_df = pd.DataFrame()
pyber_loc_df = grouped_df.loc['2019-01-01':'2019-04-29']
pyber_loc_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
#pyber_loc_df["date"]=pd.to_datetime(pyber_loc_df["date"])
##according to the pyber_data_df.info above date is already date-time
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_loc_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' 
#and get the sum of the fares for each week.

resample_fare_df = pyber_loc_df.resample("W").sum()
resample_fare_df.head()
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

resample_fare_df.plot()
plt.title("Total Fare by City Type)")
plt.ylabel("Fare($USD)")
style.use('fivethirtyeight')
plt.grid(True)
```
