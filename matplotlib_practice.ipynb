```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')
#create the labels for x and axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set the y limit between 0 and 45
plt.ylim(0,45)
#create a title
plt.title("Pyber Fare by Month")
#add the legend
plt.legend()
```

```python
#create the plot (change color and line width with first command)
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label ="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#5.1.7 scatter and bubble charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
#skill drill 5.1.7
#create the plot
plt.scatter(x_axis, y_axis)
#marker = o is circle
plt.scatter(x_axis, y_axis, marker = "o", color="red", label='Chicago')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#skill drill 5.1.7
#create the plot
#4DBEEE with quotes is the hex color for sky blue
fig, ax = plt.subplots()
#change the size
ax.scatter(x_axis, y_axis, s=y_axis)
#invert plot
plt.gca().invert_yaxis()
#marker = o is circle
ax.scatter(x_axis, y_axis, marker = "o", color="#4DBEEE", edgecolor = "black", s = [i * 5 for i in y_axis])
```

```python
#5.1.8 pie charts, used for data expressed as a percentage
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#exploded values in pie chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```
